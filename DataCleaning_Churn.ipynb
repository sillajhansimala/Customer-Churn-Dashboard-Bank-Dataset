```python
import pandas as pd
```

```python
data_churn = pd.read_excel('Churn_Modelling.xlsx')
data_churn.head()
```

```python
# Dropping RowNumber, CustomerId, and Surname
data_churn.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
# Checking for missing values
data_churn.isnull().sum()
```

```python
# Fill Geography missing with most frequent value
data_churn['Geography'] = data_churn['Geography'].fillna(data_churn['Geography'].mode()[0])

# Fill Age with median
data_churn['Age'] = data_churn['Age'].fillna(data_churn['Age'].median())

# Fill binary columns with mode
data_churn['HasCrCard'] = data_churn['HasCrCard'].fillna(data_churn['HasCrCard'].mode()[0])
data_churn['IsActiveMember'] = data_churn['IsActiveMember'].fillna(data_churn['IsActiveMember'].mode()[0])
```

```python
data_churn.isnull().sum()
```

```python
#Coverting Geography and Gender to numerical form using One-hot encodoing and label encoding.
data_churn = pd.get_dummies(data_churn, columns=['Geography'], drop_first=True)

# Label encode 'Gender'
data_churn['Gender'] = data_churn['Gender'].map({'Male': 1, 'Female': 0})
```

```python
data_churn.head()
```

```python
#Converting boolean columns(Geography_Germany & Geography_Spain) to integers(0/1)
data_churn['Geography_Germany'] = data_churn['Geography_Germany'].astype(int)
data_churn['Geography_Spain'] = data_churn['Geography_Spain'].astype(int)
```

```python
data_churn.head()
```

```python
data_churn.to_csv('cleaned_telco.csv', index=False, encoding='utf-8')
```
